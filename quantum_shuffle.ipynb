```python
def DisplayMatrix(arr):
    fig = plt.figure()
    H = np.array(arr)
    plt.xlabel("matrix x-coord")
    plt.ylabel("matrix y-cood")
    plt.imshow(H, interpolation='none')
    plt.colorbar()
    plt.show()

    
```

```python
def CountBins(length, arr):
    data = []
    for i in range(0,length):
        counts = sum(x.count(i) for x in arr)
        data.append((counts))
        #data.append(np.log(counts))

    print(data)
    #plt.hist(data, bins=6, alpha=0.5)
    plt.bar(range(1, len(data) + 1), data, tick_label=[f'{i}É›' for i in range(len(data))])
    plt.xlabel('Energy quanta')
    plt.ylabel('count')

    plt.show()
```

```python
def Logit(length,arr):
    data = []
    x = (np.arange(0.0,length,1))

    for i in range(0,length):
        counts = sum(x.count(i) for x in arr)
        data.append(np.log(counts))
        
    a,b = np.polyfit(x,data, 1)
    print(a,b)
    plt.xlabel('Energy quanta')
    plt.ylabel('Log of count')
    plt.plot(x,data,'x', color='black')
    plt.grid()
    plt.plot(x,a*x + b, color='red')
    plt.show()
```

```python
#is quanta set to zero or is it just -1

def ConductMove(sizerow, sizecol):
    rand_row_source = ran.randint(0,sizerow)
    rand_col_source = ran.randint(0,sizecol)
    
    rand_row_dest = ran.randint(0,sizerow)
    rand_col_dest = ran.randint(0,sizecol)
  
    if arr[rand_row_source][rand_col_source] > 0:
        #quanta = arr[rand_row_source][rand_col_source]
        quanta = 1
        arr[rand_row_source][rand_col_source] = arr[rand_row_source][rand_col_source] - quanta
        arr[rand_row_dest][rand_col_dest] = arr[rand_row_dest][rand_col_dest] + quanta
    
   

```

```python
#part a
rows, cols = (30, 30)
arr = [[ran.randint(1,1) for i in range(cols)] for j in range(rows)]
DisplayMatrix(arr)

iterations = 10000
for i in range(0,iterations):
    ConductMove(29,29)


DisplayMatrix(arr)
CountBins(8,arr)
Logit(5,arr)
```
